## Numpy Array and basics 
Learn about numpy

```python
import numpy as np
```

### Creating array from list

```python
arr_1d = np .array([1, 2, 3, 4, 5])
print("arr 1D:", arr_1d)


arr_2d = np .array([[1, 2, 3], [4, 5, 6]])
print(arr_2d)
```

### List Vs numpy Array

```python
python_list = [1, 2, 3]
print("python list multiplication:", python_list * 2)


np_array = np.array([1, 2, 3])   #Element wise multiplication
print("numpy array multiplication:", np_array * 2)


import time

start = time.time()
python_list = [i * 2 for i in range(1000000)]
print("\n List operator time: ", time.time() - start)



start = time.time()
np_array = np.arange(1000000) * 2
print("\nNumpy operation time:", time.time() - start)
```

### Creating Array From Scratch

```python
zeros = np.zeros((4, 3))
print(zeros)

print()

ones = np.ones((2, 5))
print(ones)

print()

full = np.full((2, 2), 7)
print(full)

print()

random = np.random.random((3, 2))
print(random)

print()

sequence = np.arange(0, 10, 2)
print(sequence)
```

### Vector, Matrix and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector:", vector)

print()

matrix = np.array([[1, 2, 3],
                    [4, 5, 6]])
print("Matrix:\n", matrix)

print()

tensor = np.array([[[1, 2], [3, 4], [5, 6], [7, 8]]])
print("Tensor:", tensor)
```

### Array Properties

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print("Shape of arr:", arr.shape)
print("Number of dimensions:", arr.ndim)
print("size of arr:", arr.size)
print("Data type of arr:", arr.dtype)
```

### Array resharing

```python
arr = np.arange(12)
print("Original Array ", arr)

reshaped_arr = arr.reshape(3, 4)
print(reshaped_arr)


flattend = reshaped_arr.flatten()
print("Flattend array:", flattend)

#ravel() returns a flattened array, but it is a view of the original array whenever possible. This means that if you modify the raveled array, it may also modify the original array. In contrast, flatten() always returns a copy of the original array, so modifying the flattened array will not affect the original array.
raveled = reshaped_arr.ravel()
print("Raveled array:", raveled)

#Transpose
transposed = reshaped_arr.T
print("Transposed array:\n", transposed)
```
