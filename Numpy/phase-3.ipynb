# Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])


print("=== Zomato Sales Data Analysis ===")
print("\nSales data shape", sales_data.shape)
print("\n Sample data for 1st 3 restau:",sales_data[:, 3:])
```

```python
#Totals sales per year
print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)

#Minimum Sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print("\nMinimum sales per year:", min_sales)

#Maximum Sales Per restuarant
max_sales = np.max(sales_data[:, 1:], axis=0)
print("\nMaximum sales per year:", max_sales)

#Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("\nAverage sales per restaurant:", avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Cumulative Sales Over Years")
plt.xlabel("Year")
plt.ylabel("Cumulative Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4])
vector2 = np.array([5, 6, 3, 9])

print("Vector Addition: ", vector1 + vector2)
print("\nVector Multiplication: ", vector1 * vector2)
print("\nDot product:", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\nAngle between vectors (radians):", angle)
```
