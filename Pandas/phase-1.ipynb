# Working With Pandas Series

## a. Creating Series

```python
pd.__version__
```

Seires through list

```python
lst = [1, 2, 3, 4, 5]
print(lst)

print()

pd.Series(lst)
```

Series Trough Numpy array

```python
arr = np.array([1, 2, 3, 4, 5])
pd.Series(arr)
```

Giving Index from our Own End

```python
pd.Series(data = ['Suman', 'Kumar', 'Dey', 'Innox1040', 'Python'], index = ['a', 'b', 'c', 'd', 'e'])
```

Series through Directory Value

```python
run_km = {'day_1': 5, 'day_2': 10, 'day_3': 15, 'day_4': 20, 'day_5': 25}
pd.Series(run_km)
```

Using repeat function along with creating a Series

```python
pd.Series(4).repeat(10)
```

```python
#Using reset function to reset the index of the series
pd.Series(4).repeat(6).reset_index(drop=True)
```

```python
s = pd.Series([10, 20]).repeat([5, 2]).reset_index(drop=True)
s
```

Accessing element

```python
s[4]
```

```python
s[6]
```

```python
s[2:4]
```

```python
s[1:-1]
```

## b) Aggregate Function on Pandas Series

```python
sr = pd.Series([1, 2, 3, 4, 5, 7, 6])

sr

sr.agg([min, max, sum])
```

## c) Series absolute function

```python
s1 = pd.Series([1, -2, -5, 6, -8, -11, 3])

s1.abs()
```

## d) Appending Series

```python
first = pd.Series([1, -2, -5, 6, -8, -11, 3])
second = pd.Series([1, 2, 3, 4, 5, 7, 6])

first, second
```

```python
NEW = pd.concat([first, second])
NEW
```

```python
NEW.reset_index(drop=True)
```
